```python
import random

def generate_seq(size, ACGT=None):
    all_nucleotides = ['A', 'C', 'G', 'T']

    if ACGT is None:
        r = random.choices(all_nucleotides, k=size)
    else:
        r = random.choices(all_nucleotides, weights=ACGT, k=size)

    final_seq_str = ''.join(r)
    return final_seq_str

random_seq = generate_seq(10000)
print(random_seq)
```

1c) Escriba un código que dada una secuencia determine 
1. la frecuencia de cada una de las bases de ADN
2. la frecuencia de cada uno de los 64 codones posibles
3. la frecuencia de cada uno de los 20 aminoácidos (para una secuencia de proteínas) y para una secuencia de ADN a partir de los codones correspondientes. 
Que el código sea capaz de mostrar los resultados en un gráfico de barras.

```python
import matplotlib.pyplot as plt

def freq_bases(dna):
    freq = {'A': 0, 'C': 0, 'G': 0, 'T': 0}
    total_bases = len(dna)
    for base in dna:
        freq[base] += 1
    for base in freq:
        freq[base] = (freq[base] / total_bases) * 100
    return freq

def freq_codons(dna):
    codon_freq = {
        'AAA': 0, 'AAC': 0, 'AAG': 0, 'AAT': 0,
        'ACA': 0, 'ACC': 0, 'ACG': 0, 'ACT': 0,
        'AGA': 0, 'AGC': 0, 'AGG': 0, 'AGT': 0,
        'ATA': 0, 'ATC': 0, 'ATG': 0, 'ATT': 0,

        'CAA': 0, 'CAC': 0, 'CAG': 0, 'CAT': 0,
        'CCA': 0, 'CCC': 0, 'CCG': 0, 'CCT': 0,
        'CGA': 0, 'CGC': 0, 'CGG': 0, 'CGT': 0,
        'CTA': 0, 'CTC': 0, 'CTG': 0, 'CTT': 0,

        'GAA': 0, 'GAC': 0, 'GAG': 0, 'GAT': 0,
        'GCA': 0, 'GCC': 0, 'GCG': 0, 'GCT': 0,
        'GGA': 0, 'GGC': 0, 'GGG': 0, 'GGT': 0,
        'GTA': 0, 'GTC': 0, 'GTG': 0, 'GTT': 0,

        'TAA': 0, 'TAC': 0, 'TAG': 0, 'TAT': 0,
        'TCA': 0, 'TCC': 0, 'TCG': 0, 'TCT': 0,
        'TGA': 0, 'TGC': 0, 'TGG': 0, 'TGT': 0,
        'TTA': 0, 'TTC': 0, 'TTG': 0, 'TTT': 0
    }
    total_codons = len(dna)
    for i in range(0, len(dna)-3):  
        codon = dna[i:i+3]
        codon_freq[codon] += 1

    for codon in codon_freq:
        codon_freq[codon] = (codon_freq[codon] / total_codons) * 100
    return codon_freq

def freq_aminoacids(prot):
    amino_acid_freq = {
        'A': 0,  # Alanine
        'R': 0,  # Arginine
        'N': 0,  # Asparagine
        'D': 0,  # Aspartic acid
        'C': 0,  # Cysteine
        'E': 0,  # Glutamic acid
        'Q': 0,  # Glutamine
        'G': 0,  # Glycine
        'H': 0,  # Histidine
        'I': 0,  # Isoleucine
        'L': 0,  # Leucine
        'K': 0,  # Lysine
        'M': 0,  # Methionine
        'F': 0,  # Phenylalanine
        'P': 0,  # Proline
        'S': 0,  # Serine
        'T': 0,  # Threonine
        'W': 0,  # Tryptophan
        'Y': 0,  # Tyrosine
        'V': 0   # Valine
    }
    total_amino_acids = len(prot)
    for amino_acid in prot: amino_acid_freq[amino_acid] += 1
    
    for amino_acid in amino_acid_freq:
        amino_acid_freq[amino_acid] = (amino_acid_freq[amino_acid]/total_amino_acids) * 100 
    return amino_acid_freq

def plot_histogram(data, data_type, xsize=5, ysize=5, xlabel=None):
    if data_type == 'nucleotides':
        freq = freq_bases(data)
    elif data_type == 'aminoacids':
        freq = freq_aminoacids(data)
    elif data_type == 'codons':
        freq = freq_codons(data)
    else:
        raise ValueError(f'Invalid data type: {data_type}, you can try with: nucleotides, aminoacids or codons')
    plt.figure(figsize=(xsize,ysize))
    plt.bar(freq.keys(), freq.values())
    if xlabel is not None:
        plt.xlabel(xlabel)
    else:
        plt.xlabel(data_type)
    plt.ylabel("Frequency")
    plt.show()


random_dna = generate_seq(10000)

plot_histogram(random_dna,"nucleotides")
plot_histogram(random_dna,"codons", xsize=15)
plot_histogram(generate_seq_protein(1000),"aminoacids")
```

```python
def orfs(chain):

    end_codons = ["TAA", "TAG", "TGA"]

    orfs_sizes= {}
    counting = False
    index=0
    s=0
    for i in range(0, len(chain)-3, 3):
        actual_codon = chain[i:i+3]
        if actual_codon in end_codons and not counting:
            index=i
            counting = True
        elif actual_codon in end_codons and counting:
            counting = False
            orfs_sizes[index] = s
            s=0
        if counting:
            s=s+1
    return orfs_sizes

def transcribe_dna(dna):
    t_dna = []
    for i in range(len(dna)):
        if dna[i] == 'T':
            t_dna.append('U')
        else:
            t_dna.append(dna[i])
    return ''.join(t_dna)

t_dna=transcribe_dna(generate_seq(1000*3))
orfs_sizes=orfs(generate_seq(1000*3))

def plot_orfs(orfs_sizes):
    plt.figure(figsize=(5,5))
    plt.plot(orfs_sizes.keys(),orfs_sizes.values())
    plt.xlabel("index of start codon")
    plt.ylabel("size of Orfs")
    plt.show()

print(f"cantidad de orfs: {orfs_sizes}")
plot_orfs(orfs_sizes)
        
            
```

```python
from collections import Counter
import pandas as pd

def base_distribution(sequence):
    count = Counter(sequence)
    total = len(sequence)
    distribution = {base: (count[base]/total)*100 for base in 'ATGC'}
    return distribution

df_results = pd.DataFrame(columns=["Sequence Type", "A", "T", "G", "C"])

rows = []
for seq in list_of_seq:
    dna_random_sequence = generate_seq(len(seq))
    coding_distribution = base_distribution(seq)
    random_distribution = base_distribution(dna_random_sequence)
    coding_row = {
        "Sequence Type": "Coding",
        "A": coding_distribution.get('A', 0),
        "T": coding_distribution.get('T', 0),
        "G": coding_distribution.get('G', 0),
        "C": coding_distribution.get('C', 0)
    }

    random_row = {
        "Sequence Type": "Random",
        "A": random_distribution.get('A', 0),
        "T": random_distribution.get('T', 0),
        "G": random_distribution.get('G', 0),
        "C": random_distribution.get('C', 0)
    }
    
    rows.append(coding_row)
    rows.append(random_row)
    
df_results = pd.DataFrame(rows, columns=df_results.columns)

print(df_results)
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X = df_results[["A", "T", "G", "C"]]
y = df_results["Sequence Type"].apply(lambda x: 1 if x == "Coding" else 0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, y_train)

y_prob = model.predict_proba(X_test)[:, 1]  

y_pred = model.predict(X_test)

print("Probabilidades de ser Coding:", y_prob)
print("Predicciones binarias:", y_pred)

accuracy = accuracy_score(y_test, y_pred)
print("Precisión:", accuracy)
```

```python
class Orf:

    def __init__(self, seq, start_index, end_index, reading_frame):
        """
        :param seq: Secuencia de ADN del ORF.
        :param start_index: Índice de inicio del ORF en la secuencia.
        :param end_index: Índice de fin del ORF en la secuencia.
        """

        self.seq = seq
        self.start_index = start_index
        self.end_index = end_index
        self.size = end_index - start_index + 1
        self.b_distr = base_distribution(seq)
        self.exon_probability = self.calculate_exon_probability()
        self.reading_frame = reading_frame


    def calculate_exon_probability(self):
        df = pd.DataFrame([self.b_distr], columns=["A", "T", "G", "C"])
        prob = model.predict_proba(df)[0, 1]
        prob = round(prob, 2)
        
        return prob



    def get_sequence(self):
        return self.seq[self.start_index:self.end_index + 1]

    def __str__(self):
        return (f"ORF: Size={self.size}, "
                f"Start Index={self.start_index}, "
                f"End Index={self.end_index}, "
                f"Sequence={self.get_sequence()}, "
                f"Exon Probability={self.exon_probability}, "
                f"Probability Distribution={self.b_distr}")

    def __repr__(self):
        return (f"Orf(seq={self.seq!r}, "
                f"start_index={self.start_index}, "
                f"end_index={self.end_index}), "
                f"Exon Probability={self.exon_probability}), "
                f"Probability Distribution={self.b_distr}")
```

```python
def obtain_all_open_reading_frames(sequence, minimal_lenght=75):
    start_codon = ["ATG"]
    end_codons = ["TAA", "TAG", "TGA"]
    all_rf = []
    my_seq_reverse = Seq.reverse_complement(sequence)
    all_orf = []


    #Obtenemos todos los marcos de lectura
    for i in range(3):
        all_rf.append(sequence[i:])
        all_rf.append(my_seq_reverse[i:])


    rf = 0
    for reading_frame in all_rf:
        rf = rf+1
        sequence_of_orf = ""
        current_in_orf = False
        start_index = 0
        for i in range(0, len(reading_frame) - 3, 3):
            codon = reading_frame[i:i + 3]
            if not current_in_orf:
                if codon in start_codon:
                    current_in_orf = True
                    sequence_of_orf = codon
                    start_index = i
            else:
                if codon in end_codons:
                    end_index = i+3
                    if end_index - start_index >= minimal_lenght:
                        current_orf = Orf(sequence_of_orf, start_index, end_index, rf)
                        all_orf.append(current_orf)
                    start_index = i+3
                    sequence_of_orf = ""
                    current_in_orf = False
                else:
                    sequence_of_orf += codon
    return all_orf

orfs_random_dna = obtain_all_open_reading_frames(my_seq)
```

```python
from scipy.stats import chisquare

def chi_square_test(distribution1, distribution2):
    dist1 = np.array(distribution1)
    dist2 = np.array(distribution2)

    dist1 = dist1 / np.sum(dist1)
    dist2 = dist2 / np.sum(dist2)
    if len(dist1) != len(dist2):
        raise ValueError("Las distribuciones deben tener el mismo tamaño.")
    chi2_statistic, p_value = chisquare(f_obs=dist1, f_exp=dist2)

    return chi2_statistic, p_value

# Distribution1 es la frecuencia de aminoacidos de una proteina basada en la traduccion de una adn random
distribution1 = list(faminoacidstraduced.values())
# Distribution2 es la frecuencia de aminoacidos de una proteina real
distribution2 = list(faminoacids.values())
print(f"Distribution 1: {distribution1} \nDistribution 2: {distribution2}", end="\n\n")
chi2_statistic, p_value = chi_square_test(distribution1, distribution2)

print("Chi-Square Statistic:", chi2_statistic)
print("P-Value:", p_value)
```

```python
from Bio import SeqIO
from Bio import Seq
import pandas as pd
import matplotlib.pyplot as plt

def read_dna_file(file):
    with open(file, 'r') as f:
        r_seq = list(SeqIO.parse(f, "fasta"))
        return r_seq[0].seq
```

```python
def obtain_orfs(f, minimal_lenght = 75, graphic=False):
    sequence = read_dna_file(f)
    all_reading_frames = obtain_all_open_reading_frames(sequence, minimal_lenght = minimal_lenght)
    my_df = pd.DataFrame({
        'READING_FRAME': [orf.reading_frame for orf in all_reading_frames],
        'START_INDEX': [orf.start_index for orf in all_reading_frames],
        'END_INDEX': [orf.end_index for orf in all_reading_frames],
        'ORF': [''.join(orf.seq) for orf in all_reading_frames],
        'SIZE': [orf.size for orf in all_reading_frames],
        'DISTRIBUTION': [orf.b_distr for orf in all_reading_frames],
        'EXON': [orf.exon_probability for orf in all_reading_frames]
    })
    
    if graphic:
        max_index = my_df['END_INDEX'].max()
        heatmap = np.zeros(max_index)
        
        for _, row in my_df.iterrows():
            start = row['START_INDEX']
            size = row['SIZE']
            exon = row['EXON']
            heatmap[start:start + size] = exon
        
        plt.figure(figsize=(10, 2))
        plt.imshow(heatmap[np.newaxis, :], cmap='bwr', aspect='auto')
        plt.colorbar(label='Probabilidad de ser un exón')
        plt.xlabel('Posición en la secuencia de ADN (START_INDEX)')
        plt.title('Mapa de calor de la probabilidad de zonas codificantes')
        plt.yticks([])
        
        plt.show()  
    return my_df
    
file = "./documents/BRCA2DNARepairAssociated.fna"
my_orfs = obtain_orfs(file, graphic=True)
```
