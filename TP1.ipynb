1a) Escriba un código que genere secuencias de ADN random de un largo preestablecido por el
usuario. Modifique el código ahora para poder controlar ademas el contenido /%) de GC.

```python
import random
def generate_seq(size, filter_nucleotides=None, prob=None):
    all_nucleotides = ['A', 'C', 'G', 'T']
    
    if filter_nucleotides is not None:
        
        nucleotides = [n for n in all_nucleotides if n not in filter_nucleotides]
        num_filter = int(size * prob)
        num_nucleotides = size - num_filter
        
        sec_filter = random.choices(filter_nucleotides, k=num_filter)
        sec_nucleotides = random.choices(nucleotides, k=num_nucleotides)
        
        final_sec = sec_filter + sec_nucleotides
        random.shuffle(final_sec)
    else:
        final_sec = random.choices(all_nucleotides, k=size)
    final_sec_str = ''.join(final_sec)
    return final_sec_str

random_seq = generate_seq(10000, ['G', 'C'], 0.5)
print(random_seq)
```

```python
from collections import Counter
import pandas as pd

def base_distribution(sequence):
    count = Counter(sequence)
    total = len(sequence)
    distribution = {base: (count[base]/total)*100 for base in 'ATGC'}
    return distribution

df_results = pd.DataFrame(columns=["Sequence Type", "A", "T", "G", "C"])

rows = []
for seq in list_of_seq:
    dna_random_sequence = generate_seq(len(seq))
    coding_distribution = base_distribution(seq)
    random_distribution = base_distribution(dna_random_sequence)
    coding_row = {
        "Sequence Type": "Coding",
        "A": coding_distribution.get('A', 0),
        "T": coding_distribution.get('T', 0),
        "G": coding_distribution.get('G', 0),
        "C": coding_distribution.get('C', 0)
    }

    random_row = {
        "Sequence Type": "Random",
        "A": random_distribution.get('A', 0),
        "T": random_distribution.get('T', 0),
        "G": random_distribution.get('G', 0),
        "C": random_distribution.get('C', 0)
    }
    
    rows.append(coding_row)
    rows.append(random_row)
    
df_results = pd.DataFrame(rows, columns=df_results.columns)

print(df_results)
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X = df_results[["A", "T", "G", "C"]]
y = df_results["Sequence Type"].apply(lambda x: 1 if x == "Coding" else 0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, y_train)

y_prob = model.predict_proba(X_test)[:, 1]  

y_pred = model.predict(X_test)

print("Probabilidades de ser Coding:", y_prob)
print("Predicciones binarias:", y_pred)

accuracy = accuracy_score(y_test, y_pred)
print("Precisión:", accuracy)
```

```python
class Orf:

    def __init__(self, seq, start_index, end_index):
        """
        :param seq: Secuencia de ADN del ORF.
        :param start_index: Índice de inicio del ORF en la secuencia.
        :param end_index: Índice de fin del ORF en la secuencia.
        """

        self.seq = seq
        self.start_index = start_index
        self.end_index = end_index
        self.size = end_index - start_index + 1
        self.b_distr = base_distribution(seq)
        self.exon_probability = self.calculate_exon_probability()


    def calculate_exon_probability(self):
        df = pd.DataFrame([self.b_distr], columns=["A", "T", "G", "C"])
        return model.predict_proba(df)[0, 1]



    def get_sequence(self):
        return self.seq[self.start_index:self.end_index + 1]

    def __str__(self):
        return (f"ORF: Size={self.size}, "
                f"Start Index={self.start_index}, "
                f"End Index={self.end_index}, "
                f"Sequence={self.get_sequence()}, "
                f"Exon Probability={self.exon_probability}, "
                f"Probability Distribution={self.b_distr}")

    def __repr__(self):
        return (f"Orf(seq={self.seq!r}, "
                f"start_index={self.start_index}, "
                f"end_index={self.end_index}), "
                f"Exon Probability={self.exon_probability}), "
                f"Probability Distribution={self.b_distr}")
```

```python
def orfs_complete(chain):
    start_codon = "ATG"
    end_codons = ["TAA", "TAG", "TGA"]
    all_orfs = []
    index=0

    while index <= len(chain) - 3:
        if start_codon == chain[index:index+3]:
            sequence = ''
            end_index = index
            for nucleotide in range(index, len(chain)-3):
                if chain[nucleotide:nucleotide+3] in end_codons:
                    sequence += chain[nucleotide:nucleotide+3]
                    end_index = nucleotide+2
                    break
                sequence += chain[nucleotide]
            orf = Orf(sequence, index, end_index)
            index=end_index
            all_orfs.append(orf)
        index+=1

    return all_orfs

orfs_random_dna = orfs_complete(my_seq)
```

```python
from scipy.stats import chisquare

def chi_square_test(distribution1, distribution2):
    dist1 = np.array(distribution1)
    dist2 = np.array(distribution2)

    dist1 = dist1 / np.sum(dist1)
    dist2 = dist2 / np.sum(dist2)
    if len(dist1) != len(dist2):
        raise ValueError("Las distribuciones deben tener el mismo tamaño.")
    chi2_statistic, p_value = chisquare(f_obs=dist1, f_exp=dist2)

    return chi2_statistic, p_value

distribution1 = list(faminoacidstraduced.values())
distribution2 = list(faminoacids.values())
print(f"Distribution 1: {distribution1} \nDistribution 2: {distribution2}", end="\n\n")
chi2_statistic, p_value = chi_square_test(distribution1, distribution2)

print("Chi-Square Statistic:", chi2_statistic)
print("P-Value:", p_value)
```
