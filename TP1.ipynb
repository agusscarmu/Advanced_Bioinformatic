1a) Escriba un código que genere secuencias de ADN random de un largo preestablecido por el
usuario. Modifique el código ahora para poder controlar ademas el contenido /%) de GC.

```python
import random
def generate_seq(size, filter_nucleotides=None, prob=None):
    all_nucleotides = ['A', 'C', 'G', 'T']
    
    if filter_nucleotides is not None:
        
        nucleotides = [n for n in all_nucleotides if n not in filter_nucleotides]
        num_filter = int(size * prob)
        num_nucleotides = size - num_filter
        
        sec_filter = random.choices(filter_nucleotides, k=num_filter)
        sec_nucleotides = random.choices(nucleotides, k=num_nucleotides)
        
        final_sec = sec_filter + sec_nucleotides
        random.shuffle(final_sec)
    else:
        final_sec = random.choices(all_nucleotides, k=size)
    final_sec_str = ''.join(final_sec)
    return final_sec_str

random_seq = generate_seq(10000, ['G', 'C'], 0.5)
print(random_seq)
```

```python
import matplotlib.pyplot as plt

def freq_bases(dna):
    freq = {'A': 0, 'C': 0, 'G': 0, 'T': 0}
    for base in dna:
        freq[base] += 1
    return freq

def freq_codons(dna):
    codon_freq = {
        'AAA': 0, 'AAC': 0, 'AAG': 0, 'AAT': 0,
        'ACA': 0, 'ACC': 0, 'ACG': 0, 'ACT': 0,
        'AGA': 0, 'AGC': 0, 'AGG': 0, 'AGT': 0,
        'ATA': 0, 'ATC': 0, 'ATG': 0, 'ATT': 0,

        'CAA': 0, 'CAC': 0, 'CAG': 0, 'CAT': 0,
        'CCA': 0, 'CCC': 0, 'CCG': 0, 'CCT': 0,
        'CGA': 0, 'CGC': 0, 'CGG': 0, 'CGT': 0,
        'CTA': 0, 'CTC': 0, 'CTG': 0, 'CTT': 0,

        'GAA': 0, 'GAC': 0, 'GAG': 0, 'GAT': 0,
        'GCA': 0, 'GCC': 0, 'GCG': 0, 'GCT': 0,
        'GGA': 0, 'GGC': 0, 'GGG': 0, 'GGT': 0,
        'GTA': 0, 'GTC': 0, 'GTG': 0, 'GTT': 0,

        'TAA': 0, 'TAC': 0, 'TAG': 0, 'TAT': 0,
        'TCA': 0, 'TCC': 0, 'TCG': 0, 'TCT': 0,
        'TGA': 0, 'TGC': 0, 'TGG': 0, 'TGT': 0,
        'TTA': 0, 'TTC': 0, 'TTG': 0, 'TTT': 0
    }
    for i in range(0, len(dna)-3):  
        codon = dna[i:i+3]
        codon_freq[codon] += 1
    return codon_freq

def freq_aminoacids(prot):
    amino_acid_freq = {
        'A': 0,  # Alanine
        'R': 0,  # Arginine
        'N': 0,  # Asparagine
        'D': 0,  # Aspartic acid
        'C': 0,  # Cysteine
        'E': 0,  # Glutamic acid
        'Q': 0,  # Glutamine
        'G': 0,  # Glycine
        'H': 0,  # Histidine
        'I': 0,  # Isoleucine
        'L': 0,  # Leucine
        'K': 0,  # Lysine
        'M': 0,  # Methionine
        'F': 0,  # Phenylalanine
        'P': 0,  # Proline
        'S': 0,  # Serine
        'T': 0,  # Threonine
        'W': 0,  # Tryptophan
        'Y': 0,  # Tyrosine
        'V': 0   # Valine
    }
    for amino_acid in prot: amino_acid_freq[amino_acid] += 1
    return amino_acid_freq

def plot_histogram(data, data_type, xsize=5, ysize=5):
    plt.figure(figsize=(xsize,ysize))
    plt.bar(data.keys(), data.values())
    plt.xlabel(data_type)
    plt.ylabel("Frequency")
    plt.show()


f_bases=freq_bases(generate_seq(100))
f_codons=freq_codons(generate_seq(100))
f_aminoacids=freq_aminoacids(generate_seq_protein(100, probabilities))

print(f"cantidad de nucleotidos: {f_bases}")
print(f"cantidad de codones: {f_codons}")
print(f"cantidad de aminoacidos: {f_aminoacids}")

# plot_histogram(f_bases,"nucleotides")
plot_histogram(f_codons,"codons", xsize=15)
plot_histogram(f_aminoacids,"aminoacids")
```

```python
def orfs(chain, type):
    if type == 'ARN':
        start_codon = "AUG"
        end_codons = ["UAA", "UAG", "UGA"]
    else:
        start_codon = "ATG"
        end_codons = ["TAA", "TAG", "TGA"]

    orfs_sizes= {}
    counting = False
    index=0
    s=0
    for i in range(0, len(chain)-3, 3):
        actual_codon = chain[i:i+3]
        if start_codon==actual_codon:
            index=i
            counting = True
        elif actual_codon in end_codons and counting:
            counting = False
            orfs_sizes[index] = s
            s=0
        if counting:
            s=s+1
    return orfs_sizes

def transcribe_dna(dna):
    t_dna = []
    for i in range(len(dna)):
        if dna[i] == 'T':
            t_dna.append('U')
        else:
            t_dna.append(dna[i])
    return ''.join(t_dna)

t_dna=transcribe_dna(generate_seq(1000))
orfs_sizes=orfs(generate_seq(1000), type='ADN')

def plot_orfs(orfs_sizes):
    plt.figure(figsize=(5,5))
    plt.plot(orfs_sizes.keys(),orfs_sizes.values())
    plt.xlabel("index of start codon")
    plt.ylabel("size of Orfs")
    plt.show()

print(f"cantidad de orfs: {orfs_sizes}")
plot_orfs(orfs_sizes)
        
            
```

Escribir un codigo que dada una secuencia de ADN determine los ORF y les asigne una probabilidad de ser (o no) un gen/exon

```python
class Orf:
    
    def __init__(self, seq, start_index, end_index):
        """
        :param seq: Secuencia de ADN del ORF.
        :param start_index: Índice de inicio del ORF en la secuencia.
        :param end_index: Índice de fin del ORF en la secuencia.
        """

        self.seq = seq
        self.start_index = start_index
        self.end_index = end_index
        self.size = end_index - start_index + 1
        self.exon_probability = self.calculate_exon_probability()

    def calculate_exon_probability(self):
        """
        Calcula la probabilidad de que una secuencia sea un exón.
    
        :param seq: Secuencia de ADN del ORF.
        :return: Probabilidad de que la secuencia sea un exón.
        """
        donor_site = "GT"
        acceptor_site = "AG"
    
        min_exon_length = 10
        max_exon_length = 3000
    

        if self.size < min_exon_length or self.size > max_exon_length:
            return 0.0
    
        donor_count = self.seq.count(donor_site)
        acceptor_count = self.seq.count(acceptor_site)
    
        atg_count = self.seq.count("ATG")
    
        # Ejemplo de cálculo de probabilidad basado en múltiples factores
        probability = 0.5
        probability += (donor_count/self.size) + (acceptor_count/self.size) + (atg_count/self.size)
    
        # Normalizar probabilidad para estar en el rango [0, 1]
        return min(probability, 1.0)


    def get_sequence(self):
        return self.seq[self.start_index:self.end_index + 1]

    def __str__(self):
        return (f"ORF: Size={self.size}, "
                f"Start Index={self.start_index}, "
                f"End Index={self.end_index}, "
                f"Sequence={self.get_sequence()}, "
                f"Exon Probability={self.exon_probability}")

    def __repr__(self):
        return (f"Orf(seq={self.seq!r}, "
                f"start_index={self.start_index}, "
                f"end_index={self.end_index}), "
                f"Exon Probability={self.exon_probability})")
```

```python
def orfs_complete(chain):
    start_codon = "ATG"
    end_codons = ["TAA", "TAG", "TGA"]
    all_orfs = []
    index=0
    
    while index < len(chain) - 3:
        if start_codon == chain[index:index+3]:
            sequence = ''
            end_index = 0
            for nucleotide in range(index, len(chain)-3):
                if chain[nucleotide:nucleotide+3] in end_codons:
                    sequence += chain[nucleotide:nucleotide+3]
                    end_index = nucleotide+2
                    break
                sequence += chain[nucleotide]
            orf = Orf(sequence, index, end_index)
            all_orfs.append(orf)
            index=end_index
        index+=1
            
    return all_orfs

orfs_random_dna = orfs_complete(my_seq)
```
